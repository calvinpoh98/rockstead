```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
import yfinance as yf
```

```python
company_names["New Code"] = [str(i) + ".SI" for i in company_names["Trading Code"]]
```

```python
x = yf.Ticker("NLC.SI")
df = x.history(period = '1y')
y = df.astype("float")
print(y.tail())
y["Close"]['2022-07-14']
```

```python
def get_returns(stock):
    #Pulling data from Yahoo. Changing the objects to float and making the date index into a column
    data = yf.Ticker(stock).history(period = '5y')
    timeframes = ["3mo","1y","2y","5y"]
    if len(data) == 0:
        return [0 for i in range(len(timeframes))]
    else:    
        data = data.astype("float")
        lastdate = data.index[-1]
    
        #finding the start date for 3mths, 1Y, 2Y, 5Y from the current date
        timedate = get_start_dates(lastdate, timeframes)

        #Slicing stock data according timeframes and appending to list according to timeframe
        gainlist = []
        lastclose = data.iloc[-1,3]
        for period in timedate:
            try:
                tempdf = data[period:]
                dividends = sum(tempdf["Dividends"])
                firstclose = tempdf['Close'][period]
                gains = (lastclose - firstclose +dividends)/firstclose
            except:
                gains = 0
            gainlist.append(round(gains,2))

    return pd.Series(gainlist)

def get_start_dates(lastdate, datelist):
    newlist = []
    #lastdate = datetime.strptime(lastdate,"%Y-%m-%d")
    for period in datelist:
        if "mo" in period:
            time = period[:-2]
            date = lastdate - relativedelta(months=int(time))
        else:
            time = period[:-1]
            date = lastdate - relativedelta(years=int(time))
        newlist.append(date)
    return newlist

get_returns("500.SI")
```

```python
company_names[["3mo","1y","2y","5y"]] = company_names["New Code"].head().apply(get_returns)
company_names.head()
```
